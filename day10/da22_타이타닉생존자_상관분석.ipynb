### 타이타닉 생존자 상관분석
- 상관계수로 두 변수간의 상관관계 유추
    - correlation coefficient
- 상관계수 값(-1 ~ 0 ~ 1)
    - 음수면 서로간의 음의 상관관계(한쪽값이 올라가면, 다른 값은 떨어짐)
    - 양수면 서로간의 양의 상관관계(한쪽값이 올라가면, 다른 값도 올라감)
    - 0.0 ~ 0.2 : 상관관계 거의 없다
    - 0.2 ~ 0.4 : 약한 상관관계가 있다
    - 0.4 ~ 0.6 : 상관관계가 있다
    - 0.6 ~ 0.8 : 강한 상관관계가 있다
    - 0.8 ~ 1.0 : 서로간의 매우 긴밀한 연관이 있다. 상관관계 최고

##### 결측치를 채워넣는 방법
- DF.describe()로 구해지는 기술통계값을 기준을 채워넣는 것이 가장 일반적
1. 평균값
2. 중앙값
3. 최소값(min), 최대값(max)
4. 최빈값 - 가장 데이터 개수가 많은 것
5. 최소빈도값 - 너무 데이터가 없어서 통계치에 반영하기 힘들 때
6. 기타

```python
dfTitanic.info()
```

```python
dfTitanic.describe()
```

```python
# 조사결과 나이대는 20대 중후반이 가장 많음
ageMedian = dfTitanic['age'].median()
```

```python
# 나이의 빈 값을 중앙값으로 채워넣음
dfTitanic['age'] = dfTitanic['age'].fillna(ageMedian)
```

```python
dfTitanic.info()
```

```python
# 정박항 결측치 입력, 대부분 S(southhampton)에서 배를 탔음
dfTitanic['embarked'].value_counts()
```

```python
dfTitanic['embarked'] = dfTitanic['embarked'].fillna('S')
```

```python
dfTitanic['embark_town'] = dfTitanic['embark_town'].fillna('Southhampton')
```

```python
dfTitanic.info()
```

```python
## 데크 결측치 제거
dfTitanic['deck'].value_counts()
```

```python
## 데크가 제일 많은 빈도로 나온것 : C데크
## 각 데크별로 빈도의 차이가 별로 없기 때문에
## C데크를 입력, 데크로 상관관계 비교는 사용하지 않는 것이 좋음
dfTitanic['deck'] = dfTitanic['deck'].fillna('C')
```

```python
dfTitanic = pd.read_excel('./data/타이타닉생존자_전처리완.xlsx')
```

```python
## 상관관계를 히트맵 시각화
def categoryAge(x):
    if x < 10: return 0
    elif x < 20: return 1
    elif x < 30: return 2
    elif x < 40: return 3
    elif x < 50: return 4
    elif x < 60: return 5
    elif x < 70: return 6
    elif x < 80: return 7
    else : return 8
```

```python
# 히트맵에서 사용하기 위해서 컬럼 재생산
dfTitanic['age2'] = dfTitanic['age'].apply(categoryAge) # 컬럼의 값을 함수를 적용해서 값을 낼 수 있음
dfTitanic['sex'] = dfTitanic['sex'].map({'male': 1, 'female':0}) # 여러번 수행하면 안됨
dfTitanic['family'] = dfTitanic['sibsp'] + dfTitanic['parch'] + 1
```

```python
dfTitanic.tail()
```

```python
## 히트맵
hData = dfTitanic[['survived', 'sex', 'age2', 'family', 'pclass', 'fare']]
## 속성정리, square= 각 히트맵 셀 정사각형
sns.heatmap(hData.astype(float).corr(), linewidths=1.0, square= False, vmax=1.0, cmap='YlGn', annot=True, annot_kws={'size':10})

plt.title('타이타닉 속성별 상관관계 히트맵')
plt.show()
```

- 상관관계는 음의 관계, 양의 관계 모두 중요. 절대값 0.4 ~ 1.0 사이의 값에 대한 속성간의 분석을 시작하면 됨
- 0.2 이하의 데이터는 분석대상에서 제외
