````python
## T-검정, 회귀분석을 위한 라이브러리 사용
from scipy import stats
from statsmodels.formula.api import ols, glm
````

````python
# 레드와인 필터링
condition = dfTotalWine['color'] == 'red'
````

````python
dfTotalWine.loc[condition, 'quality']
````

````python
# 레드와인 품질
redWineQuality = dfTotalWine[condition]['quality']
````

````python
condition2 = dfTotalWine['color'] == 'white'
````

````python
whiteWineQuality = dfTotalWine[condition2]['quality']
````

````python
# p-value가 0.05 이상이면 이 검정은 신뢰할 수 없다.
# statistic +0 redWineQuality 평균이 whiteWineQuality의 평균보다 크다
# redWine 평균이 whiteWine 평균보다 작다
stats.ttest_ind(redWineQuality, whiteWineQuality, equal_var= False)
````

````python
whiteWineQuality.value_counts()
````

````python
## 회귀분석(Regression)
## 11개의 독립변수 요소를 다 더해서 그 결과로 종속변수 quality의 값이 결정된다.
RFormula = 'quality ~ fixed_acidity+volatile_acidity+citric_acid+residual_sugar+chlorides+free_sulfur_dioxide+total_sulfur_dioxide+density+pH+sulphates+alcohol'
````

````python
## 회귀분석 결과
regressionResult = ols(RFormula, data= dfTotalWine).fit()
````

````python
## 회귀분석 요약
regressionResult.summary()
````
